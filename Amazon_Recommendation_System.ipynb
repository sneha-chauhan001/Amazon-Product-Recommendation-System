```python
df = df[pd.to_numeric(df['rating'], errors='coerce').notnull()]
df['rating'] = df['rating'].astype(float)
```

```python
df['combined_features'] = df['product_name'] + ' ' + df['category'] + ' ' + df['about_product']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim.shape
```

```python
indices = pd.Series(df.index, index=df['product_name']).drop_duplicates()
```

```python
def recommend_products(product_name, top_n=5):
    idx = indices[product_name]

    if isinstance(idx, pd.Series):
        idx = idx.iloc[0]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    product_indices = [i[0] for i in sim_scores]

    return df['product_name'].iloc[product_indices]
```

# Content-Based Recommendation System

## Overview
Content-Based Recommendation System recommends products based on the features of the products themselves. It analyzes product information such as product name, category, and description, and recommends items that are similar to the one selected by the user.

In this project, we use:
- **Product Name**
- **Category**
- **About Product**

to create a combined feature set for each product.

---

## Steps Performed

### 1. Feature Combination
The important product-related columns were merged into a single column named `combined_features`.

Purpose:
To create a complete textual representation of each product.

---

### 2. TF-IDF Vectorization
TF-IDF (Term Frequency - Inverse Document Frequency) was applied on the combined features.

Purpose:
To convert text data into numerical vectors so that the machine can process and compare products.

---

### 3. Cosine Similarity Calculation
Cosine similarity was calculated between all products.

Purpose:
To measure how similar one product is to another.

Range:
- **0** → Completely different
- **1** → Exactly similar

---

### 4. Recommendation Function
A function was created to:
- Take a product name as input
- Find its similarity with all products
- Return the top 5 most similar products

---

## Result
The model successfully recommends similar products.

Example:
If the input product is a **Television**, the output also recommends other televisions from similar brands and categories.

This shows that the content-based recommendation system is working correctly.

```python
user_similarity = cosine_similarity(user_item_filled)
```

```python
user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_filled.index,
    columns=user_item_filled.index
)
```

```python
user_similarity_df.head()
```

```python
def recommend_collaborative(user_id, top_n=5):
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:top_n+1]

    similar_users_products = user_item_filled.loc[similar_users.index]

    recommended_products = similar_users_products.mean().sort_values(ascending=False)

    return recommended_products.head(top_n)
```

## Collaborative Filtering

In this section, we implemented the **Collaborative Filtering Recommendation System**.  
Unlike content-based filtering, collaborative filtering recommends products based on the behavior and preferences of **similar users**.

### Steps performed:
1. Created a **User-Item Matrix** using `user_id`, `product_id`, and `rating`.
2. Filled missing values (`NaN`) with `0` to represent no interaction.
3. Applied **Cosine Similarity** to measure similarity between users.
4. Built a recommendation function to:
   - Find the most similar users.
   - Collect their highly rated products.
   - Recommend top products to the selected user.

### Observation:
The dataset was highly sparse, meaning most users rated only a small number of products.  
Because of this, user-user similarity was often low.

### Result:
The system successfully generated recommendations based on similar user behavior.  
However, due to sparse interactions, collaborative filtering may be less effective than content-based filtering for this dataset.

This limitation is common in real-world recommendation systems and is known as the **sparsity problem**.

```python
def hybrid_recommend(product_name, user_id, top_n=5):

    content_recs = recommend_products(product_name, top_n).tolist()

    collaborative_recs = recommend_collaborative(user_id, top_n).index.tolist()

    collaborative_names = df[df['product_id'].isin(collaborative_recs)]['product_name'].drop_duplicates().tolist()

    final_recommendations = list(dict.fromkeys(content_recs + collaborative_names))

    return final_recommendations[:top_n]
```

```python
hybrid_recommend(
    df['product_name'].iloc[123],
    df['user_id'].iloc[0]
)
```

## Hybrid Recommendation System

In this section, we implemented the **Hybrid Recommendation System** by combining both Content-Based Filtering and Collaborative Filtering.

Hybrid recommendation takes advantage of:
- **Content-Based Filtering** for personalized recommendations based on product features.
- **Collaborative Filtering** for discovery of products based on similar user behavior.

### Steps performed:
1. Generated content-based recommendations using product features such as product name, category, and description.
2. Generated collaborative recommendations using user-product-rating interactions.
3. Converted collaborative product IDs into product names.
4. Combined both recommendation lists.
5. Removed duplicate recommendations.

### Observation:
The hybrid model provided a better balance between personalization and discovery.

### Result:
Hybrid filtering produced more balanced and practical recommendations by reducing the limitations of both individual systems.

## Comparison of Recommendation Systems

| Recommendation System | Strengths | Weaknesses |
|---|---|---|
| Content-Based Filtering | Personalized, fast, uses product features | Can become repetitive |
| Collaborative Filtering | Helps discover new products, uses crowd behavior | Suffers from sparse data problem |
| Hybrid Filtering | Combines personalization and discovery, more balanced | Slightly more complex |

### Dataset Observation:
The Amazon dataset used in this project was highly sparse, meaning users interacted with very few products.  
Because of this, Content-Based Filtering performed better than Collaborative Filtering.

Hybrid Filtering gave the most balanced result overall.

## Conclusion

In this project, we implemented and compared three recommendation systems: Content-Based Filtering, Collaborative Filtering, and Hybrid Filtering.

The Content-Based model performed well because product information such as name, category, and description was rich and useful for similarity matching.

The Collaborative model worked successfully but was limited by the sparse user-product interaction data.

The Hybrid model provided the best overall performance by combining both approaches, improving both personalization and product discovery.

This project shows that Hybrid Recommendation Systems are often the most effective in real-world e-commerce platforms like Amazon.
